```python
# Librería para guardar los Trials.
import joblib

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter  # Para formatear los ejes
```

### Carga de archivo con iteraciones de optimización

```python
study = joblib.load('pickle/study_LightGBM_clean_10Fold.pkl')

df = study.trials_dataframe()[['number', 'value']]
df['number'] = df['number'] + 1

df = df.sort_values(by='value', ascending=True)

top_10 = df.nsmallest(10, 'value')[['number', 'value']]
```

### 10-Fold

```python
x = df.number
y = df.value*-1

plt.figure(figsize=(17, 7))

plt.scatter(top_10.number, top_10.value*-1, alpha=1.0, color='red', s=80)

plt.scatter(x, y, alpha=1.0, color='purple')

plt.title("Distribuciones de F1 Scores V1 [10-Fold]", fontsize=25)
plt.ylabel("F1 Score", fontsize=20)
plt.xlabel("Iteraciones", fontsize=20)
# plt.xticks(rotation=0, fontsize=17, ha="right")

plt.yticks(fontsize=17)
plt.gca().yaxis.set_major_formatter(FormatStrFormatter("%.4f"))
plt.grid()

plt.xticks([0,20,40,60,80,100])

plt.xlim([0, 101])

plt.show()
```

```python
x = df.number
y = df.value*-1

plt.figure(figsize=(17, 7))

plt.scatter(top_10.number, top_10.value*-1, alpha=1.0, color='red', s=80)

plt.scatter(x, y, alpha=1.0, color='purple')

plt.title("Distribuciones de F1 Scores V2 [10-Fold]", fontsize=25)
plt.ylabel("F1 Score", fontsize=20)
plt.xlabel("Iteraciones", fontsize=20)
# plt.xticks(rotation=0, fontsize=17, ha="right")

plt.yticks(fontsize=17)
plt.gca().yaxis.set_major_formatter(FormatStrFormatter("%.4f"))
plt.grid()

plt.xticks([0,20,40,60,80,100])

plt.ylim([0.727, 0.747])
plt.xlim([0, 101])

plt.show()
```

```python
x = top_10.number
y = top_10.value*-1

plt.figure(figsize=(17, 7))

plt.scatter(x, y, alpha=1.0, color='purple', s=100)

plt.title("Distribuciones del Top 10 de F1 Scores [10-Fold]", fontsize=25)
plt.ylabel("F1 Score", fontsize=20)
plt.xlabel("Iteraciones", fontsize=20)
# plt.xticks(rotation=0, fontsize=17, ha="right")

plt.yticks(fontsize=17)
plt.gca().yaxis.set_major_formatter(FormatStrFormatter("%.4f"))
plt.grid()

plt.xticks([80,85,90,95,100])

plt.ylim([0.7457, 0.7463])
plt.xlim([80, 101])

plt.show()
```
